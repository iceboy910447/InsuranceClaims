In this case the correlation matrix is no good indicator for which features to choose. Instead i will choose the features based on the knowledge gained within the EDA. Based on the previous exploration i have decided to drop the Region column, convert the categorical features with dummy variables.

Additionally i will convert the Claim Amount to an expected Yearly Claim Amount by dividing with Exposure

Additionally i have decided to trop ClaimNB in all frames, since it is not necessarily indicative. In only recources like https://www.kaggle.com/code/floser/use-case-claim-frequency-modeling-python this feature gets also dropped, since it is only partially mirrors the results in df_sev.

```python
def convert_and_drop(df):
    df = df.drop(columns=["Region","IDpol"])
    df = pd.get_dummies(df,columns=["Area","VehBrand","VehGas"])
    df = df.astype("float32")
    return df

df_freq = convert_and_drop(df_freq)
df_sev = convert_and_drop(df_sev)
df_total = convert_and_drop(df_total)
df_total_zwei = convert_and_drop(df_total_zwei)

df_total["ClaimAmount_per_Year"] = df_total["ClaimAmount"]/df_total["Exposure"]
df_total_zwei["ClaimAmount_per_Year"] = df_total_zwei["ClaimAmount"]/df_total_zwei["Exposure"]
df_freq.drop(columns=["ClaimAmount","ClaimNb",],inplace=True)
df_total.drop(columns=["ClaimAmount","ClaimNb","ClaimNb_new"],inplace=True)
df_total_zwei.drop(columns=["ClaimAmount","ClaimNb","ClaimNb_new"],inplace=True)

#y = df_big["ClaimAmount_per_Year"]
#X = df_big.drop(columns=["IDpol","ClaimAmount","ClaimAmount_per_Year"])
```

## Modell comparison

### Machine Learning Approaches

For the machine learning part of this notebook, i want to explore 2 different approaches. 

1. In the first approach i want to predict the yearly amount directly. To achieve this, i will use several regressor, test which is the best, and than tune the hyperparameters
2. For the second approach i wnat to predict the likelyhood that a customer will make a new claim within a year. Additionally i will train a regressor to predict the claim amount for all claims made. The yearly amount than will be the likelyhood of a claim multiplied with the probable claim amount.



```python
from sklearn.ensemble import AdaBoostRegressor, BaggingRegressor, ExtraTreesRegressor, GradientBoostingRegressor, RandomForestRegressor
from sklearn.linear_model import LinearRegression, Ridge, SGDRegressor
from sklearn.linear_model import ElasticNet, Lars, Lasso, LassoLars, OrthogonalMatchingPursuit
from sklearn.linear_model import ARDRegression, BayesianRidge

from sklearn.linear_model import HuberRegressor, RANSACRegressor, TheilSenRegressor
from sklearn.linear_model import PoissonRegressor, TweedieRegressor, GammaRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV, train_test_split
from sklearn.preprocessing import StandardScaler,RobustScaler, MinMaxScaler, PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import RFECV


### Classifier
from sklearn.ensemble import AdaBoostClassifier, BaggingClassifier, ExtraTreesClassifier, GradientBoostingClassifier, RandomForestClassifier 
from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC, LinearSVC
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import f1_score


```

### first approach

```python
X= df_total.drop(columns=["ClaimAmount_per_Year"])
y= df_total["ClaimAmount_per_Year"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
scaler = StandardScaler()
X_train= scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
def test_modell_class(Modellclass,classname):
    start_time = time.time()
    reg = Modellclass()
    reg.fit(X_train,y_train)
    train_score = reg.score(X_train,y_train)
    test_score = reg.score(X_test,y_test)
    end_time = time.time()
    print("================================================")
    print("Modell = {}".format(classname))
    print("R2 training score = {}".format(train_score))
    print("R2 test score = {}".format(test_score))
    print("Total time for training and evaluation = {}".format((end_time-start_time)))
    return reg.predict(X_test)
```

### Prepare Hyperparameters

Many regressors share the same parameters, both in name and regressors. Therefore we define the standard parameters and build dictionaries for each regressor class. This dictionaries enable us to perform a parametersearch for hyperparameter optimization using methods like RandomizedSearchCV or GridSearchCV. Alternatively BayesSearchCV can be used, but this requires a different definition

```python
#Define Standard Parameters

alpha = [0.0,0.0000001,0.000001,0.00001,0.0001,0.001,0.01,0.1,1.0,0.95,0.99]
criterion = ["squared_error", "absolute_error", "friedman_mse", "poisson"]
criterion_drei = ["gini", "entropy", "log_loss"]
criterion_zwei = ["squared_error", "friedman_mse"]
learning_rate = [0.0001,0.001,0.01,0.1,1.0],
max_depth = [4,6,8,12,16,24,32,64,128]
max_iter = [100,1000,10000,15000,100000]
n_estimators = [8,16,25,50,100,200,500]
samples = [0.2,0.5,0.8,1.0]
selection = ["cyclic","random"]

tol = [0.0000001,0.000001,0.00001,0.0001,0.001,0.01]
```

```python
#Parameter Dictionaries

adaboost_params = {"n_estimators":n_estimators,
                   "learning_rate":[1.0,0.1,0.001,0.0001],
                   "loss":["linear","square","exponential"],}
bagging_params = {"n_estimators":[2,5,10,20,50,100,200],
                  "max_samples":samples,
                  "max_features":samples,
                 }
rf_reg_params={"n_estimators":n_estimators,
                     "max_depth":max_depth,
                     "criterion":criterion,}
rf_class_params={"n_estimators":n_estimators,
                     "max_depth":max_depth,
                     "criterion":criterion_drei,}
dt_class_params= {"max_depth":max_depth,
                  "splitter":["best","random"],
                     "criterion":criterion_drei,}
kneighbor_params = {"n_neighbors":[5,10,15,20,40,80,160],
                        "weights":["uniform","distance"],
                        "algorithm":["auto","ball_tree","kd_tree","brute"],
                        "leaf_size":[15,30,60],
                        "p":[1,2],}
mlp_params = {"hidden_layer_sizes":[[100,],[100,100],[50,],[50,50],[200,],[200,200],],
              "activation":["identity","logistic","tanh","relu"],
              "solver":["lbfgs","sgd","adam"],
              "batch_size":["auto",32,64,128],
              "learning_rate":["constant","invscaling","adaptive"],
              "max_iter":[400,],}
                        
elastic_params = {"alpha":alpha,
                  "l1_ratio":[0.0,0.001,0.01,0.1,0.15,0.2,0.25,0.3,0.4,0.5,0.6,0.7,0.8],
                  "max_iter":max_iter,
                  "tol":tol,
                  "selection":selection,}
```

```python
def make_modell_dict(param_dict:dict)->dict:
    result={}
    for key, value in param_dict.items():
        newkey = "modell__{}".format(key)
        result[newkey] = value
    return result
```

#### Classifiers

```python
def optimize_classification(modell,params,n_iter=20,n_jobs=-1,cv=3):
    search = RandomizedSearchCV(modell,params,scoring = "f1",n_iter = n_iter, n_jobs=n_jobs,cv=cv)
    search.fit(X_train_class,y_train_class)
    y_train_pred = search.predict(X_train_class)
    y_test_pred = search.predict(X_test_class)
    train_score = f1_score(y_train_class,y_train_pred)
    test_score = f1_score(y_test_class,y_test_pred)
    print("f1 training score = {}".format(train_score))
    print("f1 test score = {}".format(test_score))
    print("best_params = {}".format(search.best_params_))
    return y_test_pred,search
```
